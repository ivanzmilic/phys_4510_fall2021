# Fabry-Perot Interferometer

Some reminders, it looks like this: 

### Relevant equations: 

## $ \frac{I_t}{I_i} = \frac{1}{1+ F \sin^2 {\delta/2}} $ 

where

## $ F = \frac{4R}{(1-R)^2} $

and 

## $ \delta = 2knd \cos{\theta} $

Let's go! 

```python
# let's skecth I_t vs delta first

delta = np.linspace(0,100,1001)
R = 0.5
F = 4*R / (1-R)**2.0

tr = 1.0 / (1+ F * np.sin(delta*0.5)**2.0)

plt.figure(figsize=[10,7])
plt.plot(delta,tr)
plt.xlabel("$\delta$")
plt.ylabel("$I_t / I_i$")

R = 0.25
F = 4*R / (1-R)**2.0

tr = 1.0 / (1+ F * np.sin(delta*0.5)**2.0)
plt.plot(delta,tr)

R = 0.1
F = 4*R / (1-R)**2.0

tr = 1.0 / (1+ F * np.sin(delta*0.5)**2.0)
plt.plot(delta,tr)
```

### A filter is an example of this, where we make te inner side of the cavity walls very reflective! 

```python
# wavelength of the light
llambda = np.linspace(499.5,500.5,10001) * 1E-9 # this is nm to m
# wavenumber 
k = 2.*np.pi/llambda

d = 1E-3 #thickness in m

# angle of incidence
theta = 0.0

# index 
n = 1.0
# we have mirrors on the inside! :)
R = 0.99

F = 4.*R / (1.-R)**2.0
delta = 2. * n * k * d * np.cos(theta)

tr = 1.0 / (1+ F * np.sin(delta*0.5)**2.0)

plt.figure(figsize=[10,7])
plt.plot(llambda*1E9,tr)
plt.xlabel("$\lambda$ [nm]")
plt.ylabel("$I_t / I_i$")
```

### The easiest way to change transmitance is to change d!
