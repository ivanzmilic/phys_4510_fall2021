## Wave superposition sketching:

In this notebook we want to sketch superposition of some waves. The class was online only anyway, so this should be relatively seamless!

```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib
font = {'weight' : 'normal',
        'size'   : 18}

matplotlib.rc('font', **font)
matplotlib.rc('text', usetex=True)
```

```python
# Let's start with a pretty simple example, adding two traveling perturbations (non-harmonic)

def wave1(x,t,v):
    wave1 = np.exp(-(x-v*t)**2.0)
    return wave1

def wave2(x,t,v):
    wave2 = np.exp(-(x-v*(t+2))**2.0)
    return wave2
```

```python
x_grid = np.linspace(0,10,201)
t_grid = np.linspace(0,10,201)
```

```python
# Let's make a wave that is the sum of these two:
t = 0
wave_1 = wave1(x_grid,t,1.0)
wave_2 = wave2(x_grid,t,1.0)

wave_sum = wave_1+ wave_2 # this is also a wave:
```

```python
plt.figure(figsize=[9,6])
plt.plot(x_grid,wave_1,label='$\Psi_1$')
plt.plot(x_grid,wave_2,label='$\Psi_2$')
plt.plot(x_grid,wave_sum,label='$\Psi_1+\Psi_2$')
plt.xlabel("x")
plt.ylabel("$\Psi$")
plt.legend()
```

#### How do we know this is wave? Well it is a traveling disturbance, convince yourself by plotting it in two different time instances!

```python
# To do this, let's calculate each wave on a grid, add them together and plot what we want: 

wave_1 = wave1(x_grid[:,None],t_grid[None,:],1.0)
wave_2 = wave2(x_grid[:,None],t_grid[None,:],1.0)

wave_sum = wave_1+ wave_2 # this is also a wave:
wave_sum.shape
```

```python
plt.figure(figsize=[9,6])
plt.plot(x_grid,wave_sum[:,0],label='t='+str(t_grid[0]))
plt.plot(x_grid,wave_sum[:,40],label='t='+str(t_grid[40]))
plt.xlabel("x")
plt.ylabel("$\Psi$")
plt.legend()
```

### Ok, now let's move to harmonic waves! As these are our favorites. 

```python
# Let's add two waves of different phase but same wavelength and frequency

llambda = 1.0 #
T = 0.5
k = 2.*np.pi/llambda
omega = 2.*np.pi/T
phi_1 = 0.0
phi_2 = 0.0
wave_1 = np.cos(k*x_grid[:,None]-omega*t_grid[None,:]+phi_1)
wave_2 = np.cos(k*x_grid[:,None]-omega*t_grid[None,:]+phi_2)

wave_sum = wave_1+wave_2

plt.figure(figsize=[9,6])
plt.plot(x_grid,wave_1[:,0],label='$\Psi_1$')
plt.plot(x_grid,wave_2[:,0],label='$\Psi_2$')
plt.plot(x_grid,wave_sum[:,0],label='$\Psi_1+\Psi_2$')
plt.xlabel("x")
plt.ylabel("$\Psi$")
plt.legend()
```

```python
# Let's add two waves of different phase but same wavelength and frequency
llambda = 1.0 #
T = 0.5
k = 2.*np.pi/llambda
omega = 2.*np.pi/T
phi_1 = 0.0
phi_2 = np.pi/2.0
wave_1 = np.cos(k*x_grid[:,None]-omega*t_grid[None,:]+phi_1)
wave_2 = np.cos(k*x_grid[:,None]-omega*t_grid[None,:]+phi_2)
wave_sum = wave_1+wave_2

plt.figure(figsize=[9,6])
plt.plot(x_grid,wave_1[:,0],label='$\Psi_1$')
plt.plot(x_grid,wave_2[:,0],label='$\Psi_2$')
plt.plot(x_grid,wave_sum[:,0],label='$\Psi_1+\Psi_2$')
plt.xlabel("x")
plt.ylabel("$\Psi$")
plt.legend()
```

```python
# Let's add two waves of different phase but same wavelength and frequency
llambda = 1.0 #
T = 0.5
k = 2.*np.pi/llambda
omega = 2.*np.pi/T
phi_1 = 0.0
phi_2 = np.pi
wave_1 = np.cos(k*x_grid[:,None]-omega*t_grid[None,:]+phi_1)
wave_2 = np.cos(k*x_grid[:,None]-omega*t_grid[None,:]+phi_2)
wave_sum = wave_1+wave_2

plt.figure(figsize=[9,6])
plt.plot(x_grid,wave_1[:,0],label='$\Psi_1$')
plt.plot(x_grid,wave_2[:,0],label='$\Psi_2$')
plt.plot(x_grid,wave_sum[:,0],label='$\Psi_1+\Psi_2$')
plt.xlabel("x")
plt.ylabel("$\Psi$")
plt.legend()
```

```python
# Of course, they do not have to have the same amplitude:
llambda = 1.0 #
T = 0.5
k = 2.*np.pi/llambda
omega = 2.*np.pi/T
phi_1 = 0.0
phi_2 = np.pi
wave_1 = 2.0 * np.cos(k*x_grid[:,None]-omega*t_grid[None,:]+phi_1)
wave_2 = np.cos(k*x_grid[:,None]-omega*t_grid[None,:]+phi_2)
wave_sum = wave_1+wave_2

plt.figure(figsize=[9,6])
plt.plot(x_grid,wave_1[:,0],label='$\Psi_1$')
plt.plot(x_grid,wave_2[:,0],label='$\Psi_2$')
plt.plot(x_grid,wave_sum[:,0],label='$\Psi_1+\Psi_2$',linewidth=3)
plt.xlabel("x")
plt.ylabel("$\Psi$")
plt.legend()
```

### Keep in mind that we can add the waves that do not have the same frequency and wavelength, as long as their speed is the same, that is, they satisfy the same wave equation.

For example:

```python
# Let's add two waves of different phase but same wavelength and frequency
llambda1 = 1.0 #
T1 = 0.5
k1 = 2.*np.pi/llambda1
omega1 = 2.*np.pi/T1

llambda2 = 0.5 #
T2 = 0.25
k2 = 2.*np.pi/llambda2
omega2 = 2.*np.pi/T2


phi_1 = 0.0
phi_2 = 0.0
wave_1 = np.cos(k1*x_grid[:,None]-omega1*t_grid[None,:]+phi_1)
wave_2 = np.cos(k2*x_grid[:,None]-omega2*t_grid[None,:]+phi_2)
wave_sum = wave_1+wave_2

plt.figure(figsize=[9,6])
plt.plot(x_grid,wave_1[:,0],label='$\Psi_1$',alpha=0.7)
plt.plot(x_grid,wave_2[:,0],label='$\Psi_2$',alpha=0.7)
plt.plot(x_grid,wave_sum[:,0],label='$\Psi_1+\Psi_2$',linewidth=3)
plt.xlabel("x")
plt.ylabel("$\Psi$")
plt.legend()
```

### However, most of the time we will focus on the superposition of the waves that have identical or very similar wavelength/frequency :)
